## 반올림

```python
google['Rating'] = google['Rating'].fillna(google['Rating'].mean())
```

```python
google['Rating_round']=google['Rating'].round()
```

```python
google['Rating_round'].value_counts()
```

```python

def get_category(Rating):
    cat = ''
    if Rating == 1.0:
        cat = 'Very Bad'
    elif Rating == 2.0:
        cat = 'Bad'
    elif Rating == 3.0:
        cat = 'So So'
    elif Rating == 4.0:
        cat = 'Good'
    elif Rating == 5.0:
        cat = 'Very good'
        
    return cat
```

```python
google['Rating_cat'] = google['Rating_round'].apply(lambda x : get_category(x))
google['Rating_cat'].value_counts()
```
