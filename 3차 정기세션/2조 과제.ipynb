## 3. XGBoost와 LightGBM 중 어느 것이 빠른지 비교해보자.

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib
import time

cust_df = pd.read_csv("./train.csv",encoding='latin-1')
print('dataset shape:', cust_df.shape)
cust_df.head(3)
```

```python
cust_df['var3'].replace(-999999,2, inplace=True)
cust_df.drop('ID',axis=1 , inplace=True)

X_features = cust_df.iloc[:, :-1]
y_labels = cust_df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_features, y_labels,
                                                    test_size=0.2, random_state=0)
```

### 3.1. XGBoost로 학습을 할 경우 걸린 시간을 적으시오.
XGBoost로 학습했을 때 걸린 시간 : ( 287.1 )

```python
from xgboost import XGBClassifier
import time

# 시작 시간을 설정하는 코드를 적으시오.
start_time = time.time()

xgb_clf = XGBClassifier(n_estimators=500, random_state=156)

xgb_clf.fit(X_train, y_train, early_stopping_rounds=100,
            eval_metric="auc", eval_set=[(X_train, y_train), (X_test, y_test)])

# 걸린 시간을 계산하는 코드를 적으시오.
xgb_time = time.time() - start_time

print('XGBoost 수행 시간 : {0:.1f}'.format(xgb_time))
```

### 3.2. LightGBM으로 학습을 할 경우 걸린 시간을 적으시오.
LightGBM으로 학습했을 때 걸린 시간 : ( 6.6 )

```python
from lightgbm import LGBMClassifier

# 시작 시간을 설정하는 코드를 적으시오.
start_time = time.time()

lgbm_clf = LGBMClassifier(n_estimators=500)

evals = [(X_test, y_test)]
lgbm_clf.fit(X_train, y_train, early_stopping_rounds=100, eval_metric="auc", eval_set=evals,
                verbose=True)

# 걸린 시간을 계산하는 코드를 적으시오.
lgbm_time = time.time() - start_time

print('XGBoost 수행 시간 : {0:.1f}'.format(lgbm_time))
```

## 4. 다음은 신용카드 데이터 세트이다. 이 데이터 세트의 이상치를 제거했을 때와 SMOTE방식을 사용했을 때 정밀도와 재현율 수치가 변화하였다.

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline

card_df = pd.read_csv('./creditcard.csv')
```

```python
from sklearn.model_selection import train_test_split

def get_preprocessed_df(df=None):
    df_copy = df.copy()
    df_copy.drop('Time', axis=1, inplace=True)
    return df_copy
```

```python
def get_train_test_dataset(df=None):
    df_copy = get_preprocessed_df(df)
    X_features = df_copy.iloc[:, :-1]
    y_target = df_copy.iloc[:, -1]
    X_train, X_test, y_train, y_test = \
    train_test_split(X_features, y_target, test_size=0.3, random_state=0, stratify=y_target)
    return X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = get_train_test_dataset(card_df)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import roc_auc_score

def get_clf_eval(y_test, pred):
    confusion = confusion_matrix( y_test, pred)
    accuracy = accuracy_score(y_test , pred)
    precision = precision_score(y_test , pred)
    recall = recall_score(y_test , pred)
    f1 = f1_score(y_test,pred)
    roc_auc = roc_auc_score(y_test, pred)
    
    print('오차 행렬')
    print(confusion)
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f},\
    F1: {3:.4f}, AUC:{4:.4f}'.format(accuracy, precision, recall, f1, roc_auc))
```

```python
def get_model_train_eval(model, ftr_train=None, ftr_test=None, tgt_train=None, tgt_test=None):
    model.fit(ftr_train, tgt_train)
    pred = model.predict(ftr_test)
    get_clf_eval(tgt_test, pred)
```

```python
from lightgbm import LGBMClassifier

lgbm_clf = LGBMClassifier(n_estimators=1000, num_leaves=64, n_jobs=-1, boost_from_average=False)
print('='*15+' 데이터 가공 전의 LightGBM 성능 지표 '+'='*15)
get_model_train_eval(lgbm_clf, ftr_train=X_train, ftr_test=X_test, tgt_train=y_train, tgt_test=y_test)
```

```python
import numpy as np

def get_outlier(df=None, column=None, weight=1.5):
    fraud = df[df['Class']==1][column]
    quantile_25 = np.percentile(fraud.values, 25)
    quantile_75 = np.percentile(fraud.values, 75)
    boundary = quantile_75 - quantile_25
    iqr_weight = boundary * weight
    lowest_val = quantile_25 - iqr_weight
    highest_val = quantile_75 + iqr_weight
    
    outlier_index = fraud[(fraud < lowest_val) | (fraud > highest_val)].index
    return outlier_index
```

```python
outlier_index = get_outlier(df=card_df, column='V14', weight=1.5)
print('이상치 데이터 인덱스:', outlier_index)
```

```python
def get_preprocessed_df(df=None):
    df_copy = df.copy()
    amount_n = np.log1p(df_copy['Amount'])
    df_copy.insert(0, 'Amount_Scaled', amount_n)
    df_copy.drop(['Time','Amount'], axis=1, inplace=True)
    outlier_index = get_outlier(df=df_copy, column='V14', weight=1.5)
    df_copy.drop(outlier_index, axis=0, inplace=True)
    return df_copy
```

```python
X_train, X_test, y_train, y_test = get_train_test_dataset(card_df)
print('='*15+' V14의 이상치 제거 후 LightGBM 성능 지표 '+'='*15)
get_model_train_eval(lgbm_clf, ftr_train=X_train, ftr_test=X_test, tgt_train=y_train, tgt_test=y_test)
```

### 4.1. V14의 이상치 제거에 대하여 다음의 물음에 답하시오.
1. V14의 이상치를 제거한 후 가공 전 데이터와 비교했을 때 정밀도는 어떻게 되었는가? 이상치를 제거했을때 정밀도가 더 증가했다.
2. V14의 이상치를 제거한 후 가공 전 데이터와 비교했을 때 재현율는 어떻게 되었는가? 이상치를 제거했을 때 재현율 또한 더 증가졌다.
3. 왜 그렇게 변했는지 서술하시오. 데이터 패턴에서 벗어난 이상치를 없애주었기 때문에 정밀도와 재현율이 동시에 증가하였다.

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=0)
X_train_over, y_train_over = smote.fit_sample(X_train, y_train)

lgbm_clf = LGBMClassifier(n_estimators=1000, num_leaves=64, n_jobs=-1, boost_from_average=False)
print('='*15+' SMOTE방법 사용 후 LightGBM 성능 지표 '+'='*15)
get_model_train_eval(lgbm_clf, ftr_train=X_train_over, ftr_test=X_test,
                  tgt_train=y_train_over, tgt_test=y_test)
```
